## d. Generar un modelo de regresión lineal que permita predecir el imdb_score de los títulos

### Regresión Lineal

```python
import seaborn as sns
```

```python
sns.scatterplot(x='release_year', y='imdb_score', data=inner_join_df)
```

```python
import statsmodels.formula.api as smf
```

```python
reg = smf.ols('imdb_score ~ release_year + imdb_votes + tmdb_popularity + tmdb_score', data=inner_join_df)
res = reg.fit()
print(res.summary())
```

## e.	Evaluar el modelo

Para este ejercicio relice un modelo de regresión multiple OLS, en donde tengo la intension de usar las variables cualitativas (release_year + imdb_votes + tmdb_popularity + tmdb_score) para explicar mi variable dependiente **imdb_score**. De esta forma podemos observar que el criterio de selección R_squared determina que las variables independientes es decir las Xs, explican mi modelo en 72,3% y el P_valor de dichas variables son significativas con una probabilidad al 99%.

"Nota: una forma de mejorar un poco mas el modelo seria aplicar logaritmo a la variable imdb_votes ya que esta variable cuantitativa las unidades son expresadas en miles y puede aumentar la dispersión en nuestro modelo."
